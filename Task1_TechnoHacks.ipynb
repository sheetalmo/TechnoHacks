**FINDING OUTLIERS**

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
plt.scatter(train_data['PassengerId'],train_data['Pclass'])
plt.xlabel('Passengers')
plt.ylabel('Pclass')
plt.title('Train Data')
plt.show()
```

**Indentify outliers**

```python
z_scores_x= stats.zscore(train_data['PassengerId'])
z_scores_y =stats.zscore(train_data['Pclass'])
threshold=4
outliers =(abs(z_scores_x) > threshold ) | (abs(z_scores_y) > threshold)
```

**Plotting outliers**

```python
plt.scatter(train_data['PassengerId'],train_data['Pclass'],c="pink",label='Normal Data')
plt.scatter(train_data.loc[outliers,'PassengerId'],train_data.loc[outliers,'Pclass'],c='red',label='Outliers')
plt.xlabel('PassengerId')
plt.ylabel('Pclass')
plt.title('Scatter plot with outliers')
plt.legend()
plt.show()
```

**Their is no outliers in given data.**
