```python
import numpy as np
import pandas as pd
```

```python
#separte datatype in categorical values
categorical_columns=df.select_dtypes(include=['object','category']).columns.tolist()
#separte datatype in numberical values
numberical_columns =df.select_dtypes(include=['int64','float64']).columns.tolist()

```

Print categorical and numberical column

```python
print(f"categorical_columns:\n{categorical_columns}")
```

```python
print(f"numberical_columns:\n{numberical_columns}")
```

```python
#creating a dataframe of the numberical columns
columns_used= ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
df1=df.loc[:,columns_used]
```

```python
df1.head()
```

Lets discover Statistics of dataset

**Mean**

```python
#mean() method to calaculate mean of each column
df1_mean=df1.mean()
print(f"The mean for each numerical column:\n{df1_mean}")
```

**Median**

```python
df1_median=df.median()

print(f"The median for each numerical column:\n {df1_median}")
```

**Mode**

```python
df1_mode=df.mode().iloc[0]
print(f"The mode for each numerical column:\n{df1_mode}")
```

```python
df1_std=df.std()
print(f"The standard deviation for each numerical column:\n {df1_std}")

```
